```python
#A function which can calculate the news with specific keywords during specific period in NewYorkTimes Website
from selenium import webdriver
import time
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from datetime import datetime, timedelta
def Calculate_Number_of_News_Period(startmonth,startday,startyear,searchstring): #Input of startdate and searchstring
    driver = webdriver.Firefox() #Pretends to be a Firefox browser
    url='http://query.nytimes.com/search/sitesearch/'
    driver.get(url) #This will open a Firefox window
    time.sleep(6)  ##This is necessary to give Firefox enough time to load the page
    element=driver.find_element_by_css_selector("a[data-refineby=\"U.S.\"]").click() #Choose the news' region in US
    time.sleep(3)
    element=driver.find_element_by_css_selector("a[data-refineby=\"customDateRange\"]").click()#choose specific news dates
    time.sleep(3)
    element=driver.find_element_by_xpath("//span[contains(text(),'From:')]") #Locate on where to input startdate
    Date=['Month','Day','Year']
    startmonth="%02d" % startmonth
    startday="%02d" % startday
    startyear=startyear
    DateTime=[str(startmonth),str(startday),str(startyear)]
    for i in range(3):
        path="//input[@name='from"+Date[i]+"']"
        element=element.find_element_by_xpath(path)
        element.send_keys(DateTime[i])      #Send start day/month/year to the website
        time.sleep(1)
    starttime=datetime.strptime(str(startmonth)+'/'+str(startday)+'/'+str(startyear),"%m/%d/%Y").date()
    endtime=starttime+timedelta(7)      #Enddate is one week after the startdate
    endmonth="%02d" % endtime.month
    endday="%02d" % endtime.day
    endyear=endtime.year
    element=driver.find_element_by_xpath("//span[contains(text(),'To:')]")
    Date=['Month','Day','Year']
    DateTime=[str(endmonth),str(endday),str(endyear)]
    for i in range(3):
        path="//input[@name='to"+Date[i]+"']"
        element=element.find_element_by_xpath(path)
        element.send_keys(DateTime[i])
        time.sleep(1)
    element=driver.find_element_by_xpath("//a[contains(text(),'Refine Search')]").click()
    time.sleep(3)
    element=driver.find_element_by_id("newSearchQueryTop")#Locate where to enter search keywords and send the searchstring
    element.send_keys(searchstring)
    time.sleep(2)
    element=driver.find_element_by_css_selector("button[type=\"submit\"]").click()   #Submit the search request
    time.sleep(6)
    NumberOfNews=driver.find_element_by_class_name("totalResultsCount").get_attribute('textContent')   #Getting the results number in website
    StartLoc=NumberOfNews.find('about ')
    EndLoc=NumberOfNews[StartLoc+6:].find(' ')
    Num=NumberOfNews[StartLoc+6:StartLoc+6+EndLoc]
    driver.quit() #Close the browser (avoid browser explosion)
    return Num  #Return the number of news
```

```python
Positive2011=[] #Calculte Positive News during 2011 with frequency of one week
searchstring='economy OR market recover OR boost OR positive OR rebound OR improve OR stimulate'
month=1
day=1
year=2011
endtime=datetime.strptime('12/31/2011',"%m/%d/%Y").date()   #Endtime
nexttime=datetime.strptime('1/1/2011',"%m/%d/%Y").date()  #Starttime
element=(str(month)+'/'+str(day)+'/'+str(year))
while (nexttime<endtime):
    starttime=datetime.strptime(element,"%m/%d/%Y").date()
    Num=Calculate_Number_of_News_Period(month,day,year,searchstring)
    startstring=str(month)+'/'+str(day)+'/'+str(year)
    nexttime=starttime+timedelta(7)
    month=int(nexttime.month)
    day=int(nexttime.day)
    year=int(nexttime.year)
    element=(str(month)+'/'+str(day)+'/'+str(year))
    Positive2011.append(Num)
```
